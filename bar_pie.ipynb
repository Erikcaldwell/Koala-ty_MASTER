```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data_info = pd.read_csv("even_cleaner_data.csv", low_memory=False)
```

```python
data_info.columns
```

```python
base_data = data_info[['service_request_id', 'source',
       'requested_datetime','service_name',
        'lat', 'long', 
       'address', 'district', 'House_Number',
       'Clean_Street', 'Neighborhood', 'api_address', 'census_tract']]
```

```python
base_data.columns
```

```python
mid_city = base_data[base_data["Neighborhood"]== "Mid-City"]
san_diego = base_data[base_data["Neighborhood"]== "San Diego"]
north_park = base_data[base_data["Neighborhood"]== "North Park"]
pacific_beach = base_data[base_data["Neighborhood"]== "Pacific Beach"]
city_heights = base_data[base_data["Neighborhood"]== "City Heights"]
```

```python
mid_city.columns
```

```python
mid_city1 = mid_city.dropna(how="any")
```

```python

mid_city_group = mid_city1.groupby('service_name')


count_mid_city = mid_city_group['service_name'].count()

count_mid_city
```

```python

count_chart = count_mid_city.plot(kind='bar')


count_chart.set_xlabel("Service Name")
count_chart.set_ylabel("Number of Calls")
count_chart.set_title("Mid City Service Calls")


plt.show()
plt.tight_layout()
```

```python
plt.pie(count_mid_city,autopct="%1.1f%%", shadow=True)
plt.title("Mid City Service Calls")
plt.tight_layout()
```
