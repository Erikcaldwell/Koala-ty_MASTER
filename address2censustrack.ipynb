```python
# Import Dependencies
import pandas as pd
import numpy as np
import requests
import json
from pprint import pprint
import gmaps
from config import gkey

# Access maps with unique API key
gmaps.configure(api_key=gkey)

#disabling SetWithWariningCopy error
#pd.set_option('mode.chained_assignment',None)
```

```python
# Make a reference to the books.csv file path
csv_path = "Data/get_it_done_311_requests_datasd_noself.csv"

# Read the csv and store into Pandas DataFrame
# low_memory=False is added to resolve an error that occured when importing the csv
df = pd.read_csv(csv_path, low_memory=False)
df.head(50)
```

```python
sample_df = df.sample(frac=0.1)
sample_df = sample_df.reset_index(drop=True)
sample_df.head()
```

```python
#google geocode API call, giving it lat and long and getting back clean addresses and neighborhoods names

ListHouse = []
ListStreet = []
ListNeib = []
lat = ''
lng = ''

for index, row in sample_df.iterrows():  
    lat = row["lat"]
    lat =str(lat)
    lng = row["long"]
    lng = str(lng)
    url = 'https://maps.googleapis.com/maps/api/geocode/json?latlng='+lat+','+lng+'&key='+gkey
    result = requests.get(url).json()
    ListHouse.append(result['results'][0]['address_components'][0]['long_name'])
    ListStreet.append(result['results'][0]['address_components'][1]['long_name'])
    ListNeib.append(result['results'][0]['address_components'][2]['long_name'])

```

```python
sample_df['House_Number'] = ListHouse
sample_df['Clean_Street'] = ListStreet
sample_df['Neighborhood'] = ListNeib


```
